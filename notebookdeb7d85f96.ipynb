```python
import pandas as pd

# Load the CSV file into a Pandas DataFrame
file_path = '/kaggle/input/data12/01.Data Cleaning and Preprocessing.csv'  # Replace with the actual file path
data = pd.read_csv(file_path)

# Display the first few rows to understand the data structure
print("Initial Data:")
print(data.head())

# Step 1: Filter Data
# Example: Filter rows where 'Y-Kappa' > 25 and 'ChipRate' > 16.5
# Update column names and conditions based on your dataset
if 'Y-Kappa' in data.columns and 'ChipRate' in data.columns:
    filtered_data = data[(data['Y-Kappa'] > 25) & (data['ChipRate'] > 16.5)]
    print("\nFiltered Data (Y-Kappa > 25 and ChipRate > 16.5):")
    print(filtered_data.head())
else:
    print("\nColumns 'Y-Kappa' or 'ChipRate' not found in the dataset.")

# Step 2: Handle Missing Values
# Replace missing values in a specific column (e.g., 'SulphidityL-4') with the column mean
# Update the column name based on your dataset
if 'SulphidityL-4' in data.columns:
    data['SulphidityL-4'] = data['SulphidityL-4'].fillna(data['SulphidityL-4'].mean())

# Drop rows with missing values in a specific column (e.g., 'UCZAA')
# Update the column name based on your dataset
if 'UCZAA' in data.columns:
    data = data.dropna(subset=['UCZAA'])

print("\nData after Handling Missing Values:")
print(data.head())

# Step 3: Calculate Summary Statistics
summary_stats = data.describe()
print("\nSummary Statistics:")
print(summary_stats)

# Step 4: Check for Remaining Missing Values
missing_values = data.isnull().sum()
print("\nRemaining Missing Values:")
print(missing_values)
print("Data Preview:")
print(data.head())

# Calculate summary statistics
summary_stats = data.describe()

# Display the summary statistics
print("\nSummary Statistics:")
print(summary_stats)

# Additional statistics (optional)
# Calculate median for all numerical columns
median_stats = data.median()
print("\nMedian for Numerical Columns:")
print(median_stats)

# Check for missing values
missing_values = data.isnull().sum()
print("\nMissing Values in Each Column:")
print(missing_values)


```
